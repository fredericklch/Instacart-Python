```python
# Create a subset of the first one million rows

df = ords_prods_merge[:1000000]
```

```python
# Group subset dataframe by product name

df.groupby('product_name')
```

```python
# Group subset dataframe by department id and apply aggregate function on order number to find mean values

df.groupby('department_id').agg({'user_order_number': ['mean']})
```

```python
# Using mean function rather than aggregate to find the mean value for gouped department id

df.groupby('department_id')['user_order_number'].mean()
```

```python
# Finding multiple aggregations on grouped department id

df.groupby('department_id').agg({'user_order_number': ['mean', 'min', 'max']})
```

```python
# Create user id grouping, performed aggregate function and add to new column 'max order'

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['user_order_number'].transform(np.max)
```

```python
# Create a loyalty flag for loyal customer

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Create a loyalty flag for Regular customer

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Create a loyalty flag for New customer

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
