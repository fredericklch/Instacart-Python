## Q5. From the line chart above, it seems like there's a difference in expenditure based on the hour. Specifically, the fluctuations between low and high price items are bigger from midnight to about 5am, and the highest expenditure happendu during these hours as well. On the other hand, the price range of the expenditure during day time is more narrow (i.e. seems to be between 11 to 13). 

```python
# Q6.Create new dataframe that only contains Age and n_dependants for the purpose of plotting the line chart

df_3 = small[['Age','n_dependants']]
```

```python
# Q6.Create line chart on Age and number of dependants. 

line2 = sns.lineplot(data = df_3, x = 'Age',y = 'n_dependants')
```

## Q6. At first glance, there doesn't seem to be a clear connection or trend  between age and number of dependants. The line chart does shows that the range of average number of dependants are more narrow between age 30 to 50, probably because this age group of people are more mature, stable, working and having their family. Interestingly, the range and fluctuations are quite big for people age 60 and above, this could be due to more and more elderly looking after their grandchildren.
