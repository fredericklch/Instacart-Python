## Q5. The common id between orders & products dataframe and products dataframe is the product_id. I will be able to do an inner join to combine the data.

```python
# Q5.Merging orders & prodcuts dataframe with products dataframe using inner join

df_merged_prods_in = df_ords_prods.merge(df_prods, on = 'product_id', indicator = 'exists')
```

```python
# Check the output for orders_products merged with product ID dataframe

df_merged_prods_in.head()
```

```python
df_merged_prods_in['exists'].value_counts()
```

```python
# Merging orders & products dataframe with products dataframe using outer join

df_merged_prods_out = df_ords_prods.merge(df_prods, on = 'product_id', how = 'outer')
```

```python
# Check the output for orders_products with outer merged with product ID dataframe

df_merged_prods_out.head()
```

```python
df_merged_prods_out['_merge'].value_counts()
```
