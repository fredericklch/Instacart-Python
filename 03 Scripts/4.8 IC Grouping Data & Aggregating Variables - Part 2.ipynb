```python
# Checking output of the ords_prods_merge dataframe

ords_prods_merge.head()
```

```python
# Q2. Find aggregated mean of 'user_order_number' column grouped by department_id 

ords_prods_merge.groupby('department_id').agg({'user_order_number': ['mean']})
```

## Q3. The results for the entire dataframe contains the average for all 21 departments, where as the subset only contains the average for 8 departments. Also, the highest average in the subset is department_id 16 (19.46) but in the full dataframe, we can see that department_id 21 has the highest average of 22.90. Similarly, the lowest average in the subset is department_id 17 with 11.29 but in the full dataframe, department_id 5 has the lowest average (15.22).

```python
# Q4.Create user id grouping, performed aggregate function and add to new column 'max order' to ords_prods_merge dataframe

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['user_order_number'].transform(np.max)
```

```python
# Q4.Create a loyalty flag for loyal customer

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Q4.Create a loyalty flag for Regular customer

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Q4.Create a loyalty flag for New customer

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking output on the new loyalty flags with user_id, loyalty_flag and user_order_number columns

ords_prods_merge[['user_id', 'loyalty_flag', 'user_order_number']].head(60)
```

```python
# Q5. Find aggregated mean of 'prices' column grouped by loyalty_flag 

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

## Q5. From the above table, we can see that the spending habits are different for loyal customers to that of regular and new customers. Loyal customers tended to buy products with lower prices (average of 10.37) where as new customers tended to buy the more expensive products (average 13.29), followed by regular customers (average 12.50). 

```python
# Q6.Create user id grouping, performed aggregate function on prices and add to new column 'avg_price' to ords_prods_merge dataframe

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Q6. Checking output on the new column 'avg_price'

ords_prods_merge.head(15)
```

```python
# Q6.Create a spending flag for low spender

ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Q6.Create a spending flag for High spender

ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking output on the new spending flags with user_id, spending_flag and avg_price columns

ords_prods_merge[['user_id', 'spending_flag', 'avg_price']].head(100)
```

```python
# Q7.Create user id grouping, performed aggregate function on days_since_prior_order and add to new column 'median_days' to ords_prods_merge dataframe

ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Q7. Checking output on the new column 'median_days_since_prior_order'

ords_prods_merge.head(15)
```

```python
# Q7.Create a frequency flag for Non-frequent customer

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Q7.Create a frequency flag for Regular customer

ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
# Q7.Create a frequency flag for Frequent customer

ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Q7.Checking output on the new frequency flags with user_id, frequency_flag and median_days_since_prior_order columns

ords_prods_merge[['user_id', 'frequency_flag', 'median_days_since_prior_order']].head(60)
```

```python
# Assign missing values to suspected outliers in prices

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
